```python
#parsing the chunks and convering values to int
def chunk_parse(chunk):
    chunk=chunk.strip()
    sep_chunk=chunk.split('\n')
    if len(sep_chunk)<5:
        print("skipping invalid chunk", sep_chunk)
        return None
    username=sep_chunk[0]
    posts= int(sep_chunk[1].split(' post')[0].replace(',',''))
    followers=int(sep_chunk[2].split(' follower')[0].replace('K','000').replace('.','').replace(',',''))
    following=int(sep_chunk[3].split(' following')[0].replace(',',''))
    name=sep_chunk[4]
    if len(sep_chunk)>5:
        acc_type=sep_chunk[5]
        bio="\n".join(sep_chunk[6:])
    else:
        acc_type='not Mentioned'
        bio='not defined'
    
    return {'username':username,'posts':posts,'followers':followers,'following':following,'name':name,'type':acc_type,'bio':bio}
chunk_parse(chunks[11])    
```

```python
all_chunks=[]
for chunk in chunks:
    parsed_chunks=chunk_parse(chunk)
    if parsed_chunks is not None:
        all_chunks.append(parsed_chunks)
print(all_chunks)

```

```python
import json 
data=json.dumps(all_chunks,indent=4)
print(data)
```

```python
#people with maximum no of posts:
max_posts=0
max_following=0
max_followers=0
chunk_with_max_following=None
chunk_with_max_posts=None
chunk_with_max_followers=None


for chunk in all_chunks:
    if max_posts<chunk['posts']:
        max_posts=chunk['posts']
        chunk_with_max_posts=chunk
    if max_followers<chunk['followers']:
        max_followers=chunk['followers']
        chunk_with_max_followers=chunk
    if max_following<chunk['following']:
        max_following=chunk['following']
        chunk_with_max_following=chunk
        
        
        
print(f"chunk with maximum posts:\n\n{chunk_with_max_posts}\n\n")
print(f"chunk with maximum followers:\n\n{chunk_with_max_followers}\n\n")
print(f"chunk with maximum following:\n\n{chunk_with_max_following}\n\n")
```

```python
#finding unique categories:
categories=[]
for chunk in all_chunks:
    categories.append(chunk['type'])
unique_categories=list(set(categories))
print(categories)
print(unique_categories)
    
```
