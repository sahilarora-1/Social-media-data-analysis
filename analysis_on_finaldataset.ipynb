# Testing of script on Finaldata.txt

```python
#loading data from our text file:
with open('finaldata.txt',encoding='utf-8') as file:
    data=file.read()
```

```python
chunks=data.split("\n\n")
#creating a list of chunks which have more than 3 parameters
chunks=[c for c in chunks if len(chunks)>3]
```

```python
#parsing the chunks and convering values to int
def chunk_parse(chunk):
    chunk=chunk.strip()
    sep_chunk=chunk.split('\n')
    if len(sep_chunk)<5:
        print("skipping invalid chunk", sep_chunk)
        return None
    username=sep_chunk[0]
    posts= int(sep_chunk[1].split(' post')[0].replace(',',''))
    followers=int(sep_chunk[2].split(' follower')[0].replace('K','000').replace('.','').replace(',',''))
    following=int(sep_chunk[3].split(' following')[0].replace(',',''))
    name=sep_chunk[4]
    if len(sep_chunk)>5:
        acc_type=sep_chunk[5]
        bio="\n".join(sep_chunk[6:])
    else:
        acc_type='not Mentioned'
        bio='not defined'
    
    return {'username':username,'posts':posts,'followers':followers,'following':following,'name':name,'type':acc_type,'bio':bio}
#chunk_parse(chunks[11])    to test the function and data 
```

```python
all_chunks=[]
for chunk in chunks:
    parsed_chunks=chunk_parse(chunk)
    if parsed_chunks is not None:
        all_chunks.append(parsed_chunks)
```
